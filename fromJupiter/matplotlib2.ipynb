```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary, 'r') # sona ; koyarak veya plt.show() yazarsak alttaki [<matplotlib.lines.Line2D at 0x1c6fed953d0>] gider.
```

```python
plt.plot(age, salary, 'green', label = 'salary')
plt.plot(age, salary_2, 'blue', label = 'salary2')
plt.xlabel('age')
plt.ylabel('salary')
plt.title('Age vs Salary')
plt.legend(loc = 1)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, 'green', label = 'salary')
plt.legend()
plt.subplot(1,2,2)
plt.plot(age, salary_2, 'blue', label = 'salary2')
plt.legend()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary, 'r')
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.set_title('Age vs Salary')
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].plot(age, age, 'red', label = 'A v A')
ax[0][0].set_xlabel('age')
ax[0][0].set_ylabel('age')
ax[0][0].set_title('Age vs Age')
ax[0][0].legend()
ax[0][1].plot(age, salary, 'blue', label = 'A v S')
ax[0][1].set_xlabel('age')
ax[0][1].set_ylabel('salary')
ax[0][1].set_title('Age vs Salary')
ax[0][1].legend()
ax[1][0].plot(age, salary_2, 'green', label = 'A v S2')
ax[1][0].set_xlabel('age')
ax[1][0].set_ylabel('salary2')
ax[1][0].set_title('Age vs Salary2')
ax[1][0].legend()
ax[1][1].plot(salary, salary_2, 'orange', label = 'S v S2')
ax[1][1].set_xlabel('salary')
ax[1][1].set_ylabel('salary2')
ax[1][1].set_title('Salary vs Salary2')
ax[1][1].legend()
plt.tight_layout();
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary, 'r', label = 'salary', ls = '-.', lw = 4, marker = 'o', markersize = 5)
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x = 36, color = 'green')
ax.axhline(y = 60000, ls = '--')
ax.set_xticks(age)
ax.set_xticklabels(age, rotation = 90)
ax.set_yticks(salary)
ax.set_yticklabels(salary, rotation = 30)
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.set_title('Age vs Salary')
plt.legend(loc = 4) # plt.legend(loc = (1.5,0.5))
plt.show();
```

**Slice Charts**

```python
slices = [59219, 55466, 47544, 36443, 35917]
langs = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.pie(slices, labels=langs , autopct='%1.2f%%', explode = (0, 0, .1, 0, 0), startangle = 90, shadow =True)
# autopct noktadan sonra ka hane olsun onu gsterir.
# explode sadan balayp yukar doru 1-2-3-4-5 ten birinin ayr durmasn istiyorsak kullanlr.
# startangle ilk indextekini saa gre x derece ilerde balatr.
# shadow = True glge efekti ekler.
ax.axis('equal'); # eit dalm salar yani % kalk dilimse tam o %likte ayrr.
```
