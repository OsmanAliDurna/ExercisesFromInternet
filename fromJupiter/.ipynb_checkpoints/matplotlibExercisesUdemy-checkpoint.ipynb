```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
```

```python
y = x * 2
```

```python
plt.plot(x,y)
# plt.show()
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,6)
plt.ylim(0,15)
plt.title("String Title")
plt.savefig("myFirstPlot.png")
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
x = np.arange(0,10)
```

```python
y = x * 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.set_xlim(1,2)
axes2.set_ylim(0,5)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")
axes2.plot(x,y)
```

```python
fig = plt.figure(figsize=(2,2),dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig("new_figure_2x2.png", bbox_inches="tight")
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (4,10), dpi=200)

axes[0][0].plot(x,y)
axes[0][0].set_xlim(2,6)
axes[0][1].set_xlabel("X Label")
axes[1][0].set_ylabel("Y Label")
axes[1][1].plot(a,b)
axes[1][1].set_title("**4")

#fig.set_figwidth(10)
fig.suptitle("Figure Level",fontsize=16)
fig.subplots_adjust(wspace = 1, hspace = 1)
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

axes[0].plot(x,y)
plt.tight_layout() # for tighter layout
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label = "X vs X")
ax.plot(x,x**2,label="X vs x^2")
ax.legend(loc = (1,1)) # for location of labels table. default => loc = 0
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color="red", lw=2, marker = "s", ls="--", ms=10, markerfacecolor = "blue", markeredgewidth=4, markeredgecolor = "orange")
ax.plot(x,x+1,color="#e6c017", linewidth=5, linestyle="dashdot")

# Linestype supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,x,color="purple",lw=5)
lines[0].set_dashes([5,2,10,2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color="purple",lw=2, marker = "o", ls="--", ms=5)
ax.plot(x,x+1,color="red",lw=2, marker = "+", ls="--", markersize = 20)
```
