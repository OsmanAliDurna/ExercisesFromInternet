13. Write a Pandas program to get those movies whose revenue more than 2 million and spent less than 1 million.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('movies_metadata.csv')
df['revenue'] = df['revenue'].fillna(0).astype(int)
spentLess = df[(df['revenue'] >= 2000000)] #& (df['budget'] <= 1000000)]
spentLess.head()
```

16. Write a Pandas program to display the movies (title, runtime) longer than 30 minutes and shorter than 360 minutes.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('movies_metadata.csv')
time = df[(df['runtime'] >= 30) & (df['runtime'] <= 360)]
time[["title","runtime"]]
```
