## Basic Matplotlib Commands

```python
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary);
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.figure(figsize=(8,4))
plt.plot(age, salary, label="salary")
plt.plot(age, salary_2, label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(age, salary, "r")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

## Matplotlib Object Oriented Method

```python
fig=plt.figure()
ax=fig.add_axes([0.0, 0.0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

```python
fig=plt.figure()

ax1=fig.add_axes([0.0, 0.0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
ax2=fig.add_axes([0.4, 0.1, 0.4, 0.3]) # left, bottom, width, height (range 0 to 1)

ax1.plot(age, salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age, salary_2, "g")
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary")
ax2.set_title("Salary by Age");
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "y")
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age, salary, "r")
ax[1,1].plot(age, salary_2, "y")
plt.tight_layout();
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(age, salary,"r")
ax[0].set_title("First Plot")

ax[1].plot(age, salary_2, "b")
ax[1].set_title("Second Plot")

fig.suptitle("Figure Level Title-General Title", fontsize=15)

plt.tight_layout();
```
