## scatter plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds=sns.load_dataset('diamonds')
exercise=sns.load_dataset('exercise')
flights=sns.load_dataset('flights')
geyser=sns.load_dataset('geyser')
iris=sns.load_dataset('iris')
penguins=sns.load_dataset('penguins')
planets=sns.load_dataset('planets')
mpg=sns.load_dataset('mpg')
tips = sns.load_dataset('tips')
titanic=sns.load_dataset('titanic')
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
tips.corr()
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
fig, ax = plt.subplots (1,2, figsize = (10,5), sharey=True)
sns.scatterplot(x="total_bill", y="tip", ax=ax[0], data=tips)
sns.scatterplot(x="total_bill", y="tip", hue="smoker", ax=ax[1], data=tips);
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill", y="tip", hue="smoker", palette="viridis", data=tips);
#'viridis', 'plasma', 'inferno', 'magma', 'cividis'
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill", y="tip", size="sex", data=tips);
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill", y="tip", s=200, alpha=0.5, data=tips);
```

# 2. displot and histplot()

```python
sns.displot(x="weight", aspect=2, data=mpg, kde=True);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.rugplot(x='weight',data=mpg, ax=ax[0])
sns.displot(x='weight', kde=False, data=mpg, ax=ax[1])  # figure level plot
```

```python
sns.histplot(x="weight", data=mpg, kde=False, bins=20);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.histplot(data=mpg,x='weight', kde=True, ax=ax[0])
sns.rugplot(x='weight',data=mpg, ax=ax[1]);
```

```python
sns.set(style="darkgrid")  #darkgrid, whitegrid, dark, white, ticks
sns.histplot(x="weight", data=mpg, kde=False);
```
