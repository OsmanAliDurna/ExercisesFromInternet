```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (12,6))

ax[0].plot(x, y, 'blue', ls = '--', label = 'X vs Y')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].spines['left'].set_color('blue')
ax[0].spines['bottom'].set_color('green')
ax[0].set_xticks(x)
ax[0].set_xticklabels(x, rotation = 90)
ax[0].legend()

ax[1].plot(x, y, 'red', ls = '--', label = 'X vs Z', marker = '+', markersize = 10, markevery = 5)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')
ax[1].set_xticks(x)
ax[1].set_xticklabels(x, rotation = 90)
ax[1].legend(loc = (1.1,0.5))

plt.show()
```

**PIE CHARTS**

```python
data = [23, 17, 35, 29, 12, 41]
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
```

```python
fig, ax = plt.subplots(figsize = (14, 6))
my_colors = ['orange', 'red', 'cyan', 'grey', 'green', 'indigo']
my_explode = (0, 0, 0.2, 0, 0.4, 0)
wp={'linewidth':1,'edgecolor':'aqua'} # her bir pie snrlarn izip ekillediriyor. 

ax.pie(data, labels = cars, explode = my_explode, autopct = '%1.1f%%', startangle = 90, wedgeprops = wp, shadow = True)
ax.set_title('cars')
ax.axis('equal')
ax.legend(loc = (0, 0.8)) # loc = ('center right')

plt.show()
```

**BOX PLOT**

```python
data = pd.read_csv('hist_data.csv')
```

```python
data.describe().T
```

```python
data.head()
```

```python
ages = data.Age
```

```python
from scipy.stats import iqr
```

```python
q75, q25 = np.percentile(ages, [75, 25])
```

```python
lower = q25 - 1.5 * iqr(ages)
```

```python
upper = q75 + 1.5 * iqr(ages)
```

```python
ages.min()
```

```python
lower
```

```python
q25
```

```python
ages.median()
```

```python
fig, ax = plt.subplots(figsize = (12,5))
ax.boxplot(ages)

ax.axhline(24)
ax.axvline(1.1)

ax.axhline(35)
ax.axvline(0.9)
```

**SCATTER PLOT**

```python
data = pd.read_csv('df1')
```

```python
data.head()
```

```python
data.rename(columns={'Unnamed: 0':'date'},inplace=True)
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize = (14, 8))
ax.scatter(data.A, data.D, s=39, c = 'blue', cmap='viridis', alpha = 0.6, marker = '+', label = 'Scatter Plot Example')
ax.set_xlabel('A')
ax.set_ylabel('D')
ax.set_title('Scatter')
plt.legend()
plt.show()
```

**BAR PLOT**

```python
teams = ['BJK', 'GS', 'FB', 'TS', 'ALN']
points = [64, 61, 59, 55, 49]
```

```python
fig, ax = plt.subplots(figsize = (12,7))
ax.bar(teams, points, width = 0.8, color = ["black", 'red', 'blue', 'purple', 'bisque'])
plt.suptitle('___LeaderShip___')
ax.set_title('Championship')
ax.set_xlabel('Teams')
ax.set_ylabel('Points')
ax.set_xticks(teams)
ax.set_xticklabels(teams, rotation = 45)
plt.tight_layout()
plt.show()
```
